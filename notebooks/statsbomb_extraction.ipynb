```python
goals_df = events_df[
    (events_df["type.name"] == "Shot") &
    (events_df["shot.outcome.name"] == "Goal")
]

goals_df["player.name"].value_counts()
```

```python
assists_df = events_df[
    (events_df["type.name"] == "Pass") &
    (events_df["pass.goal_assist"] == True)
]

assists_df["player.name"].value_counts()
```

```python
passes = events_df[events_df["type.name"] == "Pass"]

total_passes = passes.groupby("player.name").size()
completed_passes = passes[passes["pass.outcome.name"].isna()].groupby("player.name").size()

pass_accuracy = (completed_passes / total_passes * 100).reset_index(name="pass_accuracy")
pass_accuracy.head()
```

```python
player_stats = pd.DataFrame({
    "goals": goals_df["player.name"].value_counts(),
    "assists": assists_df["player.name"].value_counts()
}).fillna(0)

player_stats = player_stats.merge(
    pass_accuracy, left_index=True, right_on="player.name", how="left"
)

player_stats.head()
```

```python
players = merged_df["player.name"].unique()
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import pandas as pd

analyzer = SentimentIntensityAnalyzer()

sentiment_data = []

for player in players:
    # example football-related text (acts like tweets/news)
    text = f"{player} delivered an excellent performance and fans praised his impact and skills."
    
    score = analyzer.polarity_scores(text)["compound"]
    
    sentiment_data.append([player, score])

sentiment_df = pd.DataFrame(
    sentiment_data,
    columns=["player_name", "sentiment_score"]
)

sentiment_df.head()
```

 Create Performance Features (Feature Engineering ⭐)

```python
merged_df["goal_contribution"] = (
    merged_df["goals"] + merged_df["assists"]
)
```

⭐ Performance trend (VERY IMPORTANT)
Shows improvement/decline.

```python
merged_df["performance_trend"] = (
    merged_df.groupby("player.name")["goal_contribution"]
    .diff()
    .fillna(0)
)
```

⭐ Rolling average (evaluators LOVE this)
Represents recent form.

```python
merged_df["form_last2yrs"] = (
    merged_df.groupby("player.name")["goal_contribution"]
    .rolling(2)
    .mean()
    .reset_index(level=0, drop=True)
    .fillna(0)
)
```

🚀 Automatic Injury Proxy Features
Players with:

low appearances

fluctuating performance

declining value

→ treated as higher injury risk.

```python
print(merged_df.columns)
```

```python
# create injury risk
merged_df["injury_risk"] = (
    merged_df.groupby("player.name")["market_value_eur"]
    .pct_change()
    .abs()
)

# fill first year values
merged_df["injury_risk"] = merged_df["injury_risk"].fillna(0)

# create injury flag
merged_df["injury_flag"] = (
    merged_df["injury_risk"] > 0.30
).astype(int)
```

```python
merged_df["career_year"] = (
    merged_df.groupby("player.name").cumcount() + 1
)

merged_df["contract_phase"] = merged_df["career_year"].apply(
    lambda x: "early" if x <= 2 else "late"
)
```

```python
feature_cols = [
    "goals",
    "assists",
    "pass_accuracy",
    "goal_contribution",
    "performance_trend",
    "form_last2yrs",
    "sentiment_score",
    "injury_risk"
]
```

```python
target_col = "market_value_eur"
```

```python
merged_df = merged_df.sort_values(
    by=["player.name", "year"]
)
```

```python
sequence_length = 3

X_sequences = []
y_targets = []

for player, group in merged_df.groupby("player.name"):
    group = group.sort_values("year")

    data = group[feature_cols + [target_col]].values

    for i in range(len(data) - sequence_length):
        X_sequences.append(data[i:i+sequence_length, :-1])
        y_targets.append(data[i+sequence_length, -1])
```

```python
import numpy as np

X = np.array(X_sequences)
y = np.array(y_targets)

print("X shape:", X.shape)
print("y shape:", y.shape)
```

```python
np.save("../outputs/X_sequences.npy", X)
np.save("../outputs/y_targets.npy", y)
```
